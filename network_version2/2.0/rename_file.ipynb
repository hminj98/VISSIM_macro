```python
"""파일명 변경 코드 """
import os

# 시작 번호 값
start_num = 37

folder_path = r"C:\VISSIM_Workspace\network_new\network_version2\network_version2.1\network01_mer\vehicle_output"
excel_files = sorted([f for f in os.listdir(folder_path) if f.endswith(".xlsx")], key=lambda x: int(x.split("_")[0]))

for i in range(len(excel_files)):
    old_path = os.path.join(folder_path, excel_files[i])
    print("기존경로 : ",old_path)
    new_file_name = str(start_num+i)+"_time_output.xlsx"
    new_path = os.path.join(folder_path, new_file_name)
    print("새로운 파일 경로 : ", new_path)
    os.replace(old_path, new_path)
```
