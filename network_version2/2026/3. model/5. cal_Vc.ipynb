```python
import numpy as np
import os

folder_path = r"E:\2026년\1. 초장대 K-지하고속도로 인프라 안전 및 효율 향상 기술 개발\VISSIM_Workspace\시나리오 분석\시나리오(140-유고X)"
mer_list = sorted([f for f in os.listdir(folder_path) if f.endswith(".mer")])

speed_list = []

for filename in mer_list:
    file_path = os.path.join(folder_path, filename)
    print("작업파일 :", file_path)

    with open(file_path, "r", encoding="utf-8", errors="ignore") as file:
        lines = file.readlines()

    data_start_idx = None
    for j, line in enumerate(lines):
        if "Measurem." in line:
            data_start_idx = j
            break
    if data_start_idx is None:
        continue

    columns = [col.strip() for col in lines[data_start_idx].strip().split(";")]

    try:
        v_idx = columns.index("v[km/h]")
        t_idx = columns.index("t(Entry)")
    except ValueError:
        continue

    for line in lines[data_start_idx + 1:]:
        parts = line.strip().split(";")
        if len(parts) <= max(v_idx, t_idx):
            continue
        try:
            t_val = float(parts[t_idx])
            v_val = float(parts[v_idx])
            if t_val != -1.0 and t_val <= 12600:
                speed_list.append(v_val)
        except ValueError:
            continue

percentile_speed = np.percentile(speed_list, 85)
print("percentile_speed :", percentile_speed)
```
