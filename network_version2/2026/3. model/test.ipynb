```python
import numpy as np

#현재 차로 수
lane = 2

# 유고 발생 전후 차로수 차이
accident_lane_cnt = 0

# 유고 발생 시간
accident_start_time = 3000

# 유고 해제 시간
accident_finish_time = 3300

#유고 지속 시간
accident_range_time = (accident_finish_time - accident_start_time)/60

# 동적 포화도
s_before = 0.220909

# 중차량 비율
hv_rate = 0.061728



# 종단 경사
lane_gradient = -0.07

# 진입램프 인근 여부
r_on = 0 # 맞으면 1, 아니면 0
r_off = 0 # 맞으면 1, 아니면 0

# 유고 직전 stvm(인접)
before_stvm = 0.026251

# 유고 직후 초기 stvm(인접)
after_stvm = -0.056702

# 포화도-차로 감소 상호작용
a = accident_lane_cnt * s_before
print("a : ", a)
b = hv_rate * lane_gradient
print("b : ", b)
c = accident_lane_cnt * r_on
print("c : ", c)
log_D = accident_lane_cnt + accident_range_time + s_before + hv_rate + lane_gradient + r_on + r_off + s_before*s_before + before_stvm + (after_stvm-before_stvm) + a + b + c+2

D = np.exp(log_D)

print(log_D)
print(D)
```
