```python
import win32com.client
import time

Vissim = win32com.client.Dispatch("Vissim.Vissim")

net_path = r"E:\2026년\1. 초장대 K-지하고속도로 인프라 안전 및 효율 향상 기술 개발\VISSIM_Workspace\화성~서울(유고0)\가감차로 반영0\서울방향만\1. 유고-3000(유고2차로-서비스수준A,C,E)_260225\서비스 C\화성~서울(유고2차로_서비스C)_260225.inpx"
lay_path = r"E:\2026년\1. 초장대 K-지하고속도로 인프라 안전 및 효율 향상 기술 개발\VISSIM_Workspace\화성~서울(유고0)\가감차로 반영0\서울방향만\1. 유고-3000(유고2차로-서비스수준A,C,E)_260225\서비스 C\화성~서울(유고2차로_서비스C)_260225.layx"

TIME_FROM = 3000
TIME_TO_LIST = [3300, 3900, 4800, 5700, 6600] # 5분, 15분, 30분, 45분, 60분

Vissim.LoadNet(net_path)
Vissim.LoadLayout(lay_path)
time.sleep(1)

Vissim.Graphics.CurrentNetworkWindow.SetAttValue("QuickMode", 1)

# =================================================
# 시뮬레이션 기본 설정 (필수)
# =================================================
sim = Vissim.Simulation
sim.SetAttValue("SimRes", 10)
sim.SetAttValue("SimPeriod", 13200) # 시뮬레이션 분석 시간
sim.SetAttValue("NumRuns", 5) # 시뮬레이션 반복 횟수
#sim.SetAttValue("RandSeed", 2)


rsas = Vissim.Net.ReducedSpeedAreas.GetAll()
stop_rsas = [rsa for rsa in rsas if rsa.AttValue("Name") == "정지"]

for TIME_TO in TIME_TO_LIST:

    print(f"\n🚧 Scenario start | TIME_TO = {TIME_TO}")

    # -----------------------------
    # 시뮬레이션 초기화
    # -----------------------------

    # RSA 시간 설정

    for rsa in stop_rsas:
        rsa.SetAttValue("TimeTo", TIME_TO)
        rsa.SetAttValue("TimeFrom", TIME_FROM)


     # ▶ 시뮬레이션 초기화
    sim.Stop()

    # ▶ 연속 실행
    sim.RunContinuous()

    sim.Stop()
    print(f"✅ Scenario finished | TIME_TO = {TIME_TO}")

# 종료
Vissim.Exit()
print("\n🎉 All scenarios completed.")
```
