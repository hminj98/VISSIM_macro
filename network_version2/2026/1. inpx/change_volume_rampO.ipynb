```python
"""
    교통량 변경 자동화 코드 (RunContinuous 버전)
"""

import win32com.client
import time

Vissim = win32com.client.Dispatch("Vissim.Vissim")

net_path = r"E:\2026년\1. 초장대 K-지하고속도로 인프라 안전 및 효율 향상 기술 개발\VISSIM_Workspace\화성~서울\기초분석\2차 분석\2. 화성~서울(유입램프 1개)\화성~서울(유입램프 1개)_260127.inpx"
lay_path = r"E:\2026년\1. 초장대 K-지하고속도로 인프라 안전 및 효율 향상 기술 개발\VISSIM_Workspace\화성~서울\기초분석\2차 분석\2. 화성~서울(유입램프 1개)\화성~서울(유입램프 1개)_260127.layx"

TARGET_NAME = ["main", "main2"]
NEW_VOLUME = [1200, 2700, 4400]

RAMP_TARGET_NAME  = ["ramp1-1", "ramp2-1"]
#RAMP_TARGET_NAME  = ["ramp1-1", "ramp1-2", "ramp2-1", "ramp2-2", "ramp2-3"]
NEW_RAMP_VOLUME = [500, 750, 1200]  # 시나리오별

# =================================================
# VISSIM 로드
# =================================================
Vissim.LoadNet(net_path)
Vissim.LoadLayout(lay_path)

# =================================================
# 그래픽 최소화 (속도 핵심)
# =================================================

Vissim.Graphics.CurrentNetworkWindow.SetAttValue("QuickMode", 1)


#Vissim.Graphics.SetAttValue("Visible", False)


# =================================================
# 시뮬레이션 기본 설정
# =================================================
sim = Vissim.Simulation
sim.SetAttValue("SimRes", 10)
sim.SetAttValue("SimPeriod", 13200)
sim.SetAttValue("RandSeed", 1)

# =================================================
# Vehicle Inputs
# =================================================
vvs = Vissim.Net.VehicleInputs.GetAll()

# =================================================
# 시나리오 루프
# =================================================
for i, main_vol in enumerate(NEW_VOLUME):

    ramp_vol = NEW_RAMP_VOLUME[i]

    print(f"\n🚧 Scenario start")
    print(f"   - Main volume : {main_vol}")
    print(f"   - Ramp volume : {ramp_vol}")

    # ▶ 교통량 설정
    for vv in vvs:
        name = vv.AttValue("Name")

        if name in TARGET_NAME:
            vv.SetAttValue("Volume(1)", main_vol)

        if name in RAMP_TARGET_NAME:
            vv.SetAttValue("Volume(1)", ramp_vol)

    # ▶ 시뮬레이션 초기화
    sim.Stop()

    # ▶ 연속 실행
    sim.RunContinuous()

    sim.Stop()
    print("✅ Scenario finished")

# =================================================
# 종료
# =================================================
Vissim.Exit()
print("\n🎉 All scenarios completed.")
```
