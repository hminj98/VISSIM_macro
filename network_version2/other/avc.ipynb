```python
import pandas as pd
import requests

save_path = r"C:\Users\(주)내일이엔시 도로교통안전연구소\Desktop\VDS원시\기흥IC~양재IC(AVC).csv"

url = "https://data.ex.co.kr/openapi/avcinfo/avcOg15DataList"

params = {
    "key": "7624751987",
    "type": "json",           # json or xml
    "totlDates": "20251101",   # 집계일자 (YYYYMMDD)
}

response = requests.get(url, params=params)
response.raise_for_status()

data = response.json()

item = data["list"]

df = pd.DataFrame(item)

df = df[df["routeNm"] == "경부선"]
filtered_df =  df[df["cnznNm"].str.contains("양재IC|금토JC|대왕판교IC|판교JC|판교IC|서울TG|신갈JC|수원신갈IC|기흥IC", na=False) & ~df["cnznNm"].str.contains("기흥동탄IC|서초IC", na=False) & (df["routeNm"] == "경부선")]

filtered_df.to_csv(save_path, index=False, encoding="utf-8-sig")
```
