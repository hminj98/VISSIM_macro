```python
import Service
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install())
)

driver.get("https://data.ex.co.kr/portal/fdwn/view?type=VDS&num=16&requestfrom=dataset")

select_values = {
    "deptName" : "3           ",
    "regionName" : "KR-41       ",
    "gender" : "0           ",
    "userAgeName" : "20          ",
    "pouName" : "0           ",
    "ui_datepicker_monthid" : "9",
    "ui_datepicker_yearid" : "1998"
}
for select_id, value in select_values.items():
    try:
        Select(
            WebDriverWait(driver,10).until(
                EC.presence_of_element_located((By.ID, select_id))
            )
        ).select_by_value(value)
    except TimeoutException:
        print(f"{select_id} 요소를 찾을 수 없음")








```
