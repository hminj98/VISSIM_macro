```python
import time
import os
from datetime import datetime, timedelta
import pandas as pd
import requests
save_path = rf"C:\Users\(주)내일이엔시 도로교통안전연구소\Desktop\프로젝트\2026년\1. 초장대 K-지하고속도로 인프라 안전 및 효율 향상 기술 개발\2. 원시데이터\AVC 데이터\기흥IC~양재IC(AVC).csv"

url = "https://data.ex.co.kr/openapi/avcinfo/avcOg15DataList"
key = "7624751987"

start_date = datetime(2025, 8, 1)
end_date   = datetime(2025, 9, 30)

all_df = []
current_date = start_date

while current_date <= end_date:
    date_str = current_date.strftime("%Y%m%d")
    print(f"수집 중: {date_str}")

    params = {
        "key": key,
        "type": "json",
        "totlDates": date_str
    }

    try:
        response = requests.get(url, params=params, timeout=30)
        response.raise_for_status()

        data = response.json()

        if "list" not in data:
            print(f"데이터 없음: {date_str}")
            current_date += timedelta(days=1)
            continue

        df = pd.DataFrame(data["list"])

        df = df[
            (df["routeNm"] == "경부선") &
            (df["cnznNm"].str.contains(
                "서초IC|양재IC|금토JC|대왕판교IC|판교JC|판교IC|서울TG|신갈JC|수원신갈IC|기흥IC|동탄JC",
                na=False
            ))
        ]

        all_df.append(df)

    except Exception as e:
        print(f"에러 발생 ({date_str}): {e}")

    current_date += timedelta(days=1)
    time.sleep(0.3)

# 안전 concat
if all_df:
    final_df = pd.concat(all_df, ignore_index=True)
else:
    final_df = pd.DataFrame()

final_df.to_csv(save_path, index=False, encoding="utf-8-sig")

print("완료!")
```
