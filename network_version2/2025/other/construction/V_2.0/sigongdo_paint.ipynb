```python
import os
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

# 처리할 폴더 경로
folder_path = r"C:\Users\(주)내일이엔시 도로교통안전연구소\Desktop\급경사\시공도\vds 데이터\이상치 제거\20~24년\11월\평일\화수목\pivot"

# 색상 정의
green_fill = PatternFill(start_color="00B050", end_color="00B050", fill_type="solid")   # 초록
orange_fill = PatternFill(start_color="FFC000", end_color="FFC000", fill_type="solid")  # 주황
red_fill = PatternFill(start_color="FF0000", end_color="FF0000", fill_type="solid")      # 빨강

# 폴더 내 모든 .xlsx 파일 리스트
files = [f for f in os.listdir(folder_path) if f.endswith(".xlsx")]

print("총 처리 파일 수:", len(files))

for file_name in files:
    file_path = os.path.join(folder_path, file_name)
    print("처리 중:", file_name)

    # 엑셀 파일 불러오기
    wb = load_workbook(file_path)

    # 시트 1, 2만 처리
    for sheet_name in wb.sheetnames[:2]:
        ws = wb[sheet_name]

        for row in ws.iter_rows(min_row=2):  # 첫 행 제외
            for cell in row:
                if cell.col_idx == 1:  # 첫 열 제외
                    continue

                try:
                    v = float(cell.value)
                except:
                    continue  # 숫자 아니면 패스

                if v >= 80:
                    cell.fill = green_fill
                elif v >= 50:
                    cell.fill = orange_fill
                else:
                    cell.fill = red_fill

    # *** 기존 파일에 덮어쓰기 ***
    wb.save(file_path)
    print(" → 덮어쓰기 완료:", file_name)

print("모든 파일 처리 완료!")
```
