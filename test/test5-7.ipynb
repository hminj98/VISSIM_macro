```python
import numpy as np
import datetime

"""시뮬레이션 결과파일에서 데이터 가져오기"""
import sqlite3
import pandas as pd
import win32com.client  # Windows COM 인터페이스 사용
import os
import xml.etree.ElementTree as ET
####################### 입력 값 ############################
siml_count = 10
# 폴더 경로 설정
folder_path = r"C:\VISSIM_Workspace\network_test"  # 폴더 경로 입력 받아야함 추후에 함수에 파라미터로 처리 예정
results_path = folder_path + "\\" +"network01" + ".results"
results_files = sorted([f for f in os.listdir(results_path) if f.endswith(".db")], key=lambda x: int(x.split(".")[0]))
df_list = []


# 분석할 수치 컬럼들
for i in range(siml_count):
    # 데이터베이스 연결
    db_path = os.path.join(results_path, results_files[i])
    simrun = db_path.split("\\")[-1].split(".")[0]
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # 특정 테이블의 데이터 조회
    table_name = "DATACOLLECTIONMEASUREMENT_EvaluationTimeIntervalClass"
    query = f"SELECT * FROM {table_name};"  # 전체 데이터 조회

     # 데이터 읽기
    df = pd.read_sql_query(query, conn)
    #display(df.head(100))
    conn.close()
    df_list.append(df)

# 여러 개의 데이터프레임을 하나로 합친 후, 동일 위치(행, 열)의 평균 계산
df_avg = pd.concat(df_list).groupby(level=0).mean()
#.groupby(level=0).mean() 실행 시 문자형 데이터는 잘리기 때문에
df_avg[["ARG_TIMEINTERVAL", "OBJECT_ID"]] = df_list[0][["ARG_TIMEINTERVAL", "OBJECT_ID"]]
display(df_avg.head(10))
```
